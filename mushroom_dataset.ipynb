```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv('/content/mushrooms.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
for col in df:
    print(col,'({})'.format(df[col].nunique()), ":")
    print(df[col].value_counts())
```

```python
df.drop('veil-type',axis=1, inplace=True)
```

```python
df['class'] = df["class"].map({ 'e': 1, 'p': 0})
df.head(2)
```

```python
#convert categorical values in to numeric
df_features=df.drop('class',axis=1)
df_features=pd.get_dummies(df_features, drop_first=True)
```

```python
# split the data in to independent features and target feature
X=df_features
y=df['class']

# split the data in to X_train,X_test, y_train and y_test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0, test_size=0.2)
```

```python
X.head()
```

### fit the model with Logistic Regressor



```python
from sklearn.linear_model import LogisticRegression
class1=LogisticRegression()
class1.fit(X_train, y_train)
```

```python
y_pred= class1.predict(X_test)
```

```python
# compare the predicted values with actual values(y_test)
from sklearn.metrics import accuracy_score
accu_score= accuracy_score(y_test, y_pred)
print( 'the accuracy score=' , (accu_score))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Train the model with KNN classifier


```python
from sklearn.neighbors import KNeighborsClassifier
class2=KNeighborsClassifier(n_neighbors=5, metric= 'euclidean')
class2.fit(X_train, y_train)
```

```python
y_preds= class2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_preds))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_preds)
print(cm)
```

```python
print(classification_report(y_test, y_preds))
```

Accuracy_score for the Logistic regression  and KNN classifier is 1. It seems data is clean, hence performance of the model is good.
